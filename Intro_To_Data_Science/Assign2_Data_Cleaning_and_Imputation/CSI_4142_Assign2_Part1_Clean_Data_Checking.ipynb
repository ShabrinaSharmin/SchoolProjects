## Error no:1 (Data type checking)

### **Cell 1: Test Name and description**
  * Error name: **Data Type Error**
  * Error description: This test ensures that each column has the correct type of data. Data type miss-match can lead to errors in data manipulation and thus can results in an ineffective analytical result.

### **Cell 2: Parameters for checker**

```python
## parameter for data type checker
expected_types = {
    'Transaction ID': 'object',  # Expected to be string, which is 'object' in pandas
    'Item': 'object',            # Expected to be string
    'Quantity': 'Int64',         # Expected to be integer
    'Price Per Unit': 'float64', # Expected to be float
    'Total Spent': 'float64',    # Expected to be float
    'Payment Method': 'object',  # Expected to be string
    'Location': 'object',        # Expected to be string
    'Transaction Date': 'datetime64[ns]'  # Expected to be datetime
}
```

### **Cell 3: Checker code**

```python
## function to check type
def check_data_types(df, expected_types):
  error_type ={}
  for col, expected_type in expected_types.items():
    actual_type = df[col].dtype
    if str(actual_type) != expected_type:
      error_type[col] = f"Expected: {expected_type}, Found: {actual_type}"

  return error_type


## func to display type error
def display_type_error(data_type):
  if data_type:
    for column, error in data_type.items():
        print(f"Data type mismatch in column '{column}': {error}")
  else:
    print("All column data types are correct.")

data_type_check = check_data_types(data_dirty_cafe_sales, expected_types)
display_type_error(data_type_check)
```

### **Cell 4: Report of findings**

For the dataset, the columns **' Quantity '**  **' Price Per Unit '**, and **' Total Spent '** should be in numerical form; whereas we found them to be of object type (please, refer to the output of the previous cell). Therefore, indicating the necessity of pre processing some columns for effective and meaningful analysis of the dataset.

## Error no:2 (Range checking)

### **Cell 1: Test Name and description**
  * Error name: **Range checking Error**
  * Error description: This test ensures that each column has value that corresponds to a specific attribute is within the range. This error checking will display if the specified attribute has any out of bound. Here, any column that has Unknown/Error/Empty value has been considered as out of bound.

### **Cell 2: Parameters for checker**

```python
# #data pre-process
df = data_dirty_cafe_sales.copy()

#convert col 'price per unit' to numeric(float)
df['Price Per Unit']  = pd.to_numeric(df['Price Per Unit'], errors='coerce')

expected_menu_prices = {
    "Coffee": 2,
    "Tea": 1.5,
    "Sandwich": 4,
    "Salad": 5,
    "Cake": 3,
    "Cookie": 1,
    "Smoothie": 4,
    "Juice": 3
}
 #Item for which want to check the range
item_to_Check = 'Tea'
```

### **Cell 3: Checker code**

```python
#func to check if prices for menu items are in the correct range
def check_menu_prices(row, expected_menu, item):
  item_in_row = row['Item']
  expected_item_price = expected_menu.get(item)
  actual_price_per_unit = row['Price Per Unit']
  #check if actual and expected prices are same
  if item_in_row == item:
    if expected_item_price != actual_price_per_unit or actual_price_per_unit <= 0:
      return f"Expected: {expected_item_price}, Found: {actual_price_per_unit}"
  return None

 #apply the func check_menu_prices
df['Price Check'] = df.apply(lambda row: check_menu_prices(row, expected_menu_prices, item_to_Check), axis=1)

# Filter out only mismatched prices
mismatched_prices = df[df['Price Check'].notna()]

# Display findings
print("Mismatched Prices:")
print(mismatched_prices[['Item', 'Price Per Unit', 'Price Check']])
if len(mismatched_prices) > 0:
  print(f"There are {len(mismatched_prices)} mismatched prices for the Item {item_to_Check}")
else:
  print("No mismatched prices found.")

```

### **Cell 4: Report of findings**
From the output of our range checking, we have identified mismatched entries for the "Item" **Tea**. Here, "mismatched" refers to prices that are either less than, greater than the expected price of $1.5, or where the price data is missing altogether. The checker also identifies any negative values, as such prices are not feasible and indicate data entry errors.

The output provides a snapshot of 66 rows showing discrepancies, predominantly with "NaN" as the unit price for Tea. This highlights a significant issue with data entries, suggesting that many records are incomplete or improperly recorded. These findings underscore the necessity for a thorough data cleaning process to prepare the dataset for effective analysis.

This data cleaning will involve confirming the source of these NaN values and correcting them where possible, or deciding on an appropriate handling strategy such as imputation or exclusion, depending on the analysis needs and data integrity standards.

## Error no:4 (Consistency errors)


### **Cell 1: Test Name and description**
  * Error name: **Consistency Error**
  * Error description: This error identifies discrepancies between the recorded "Total Spent" and the product of "Price Per Unit" and "Quantity" for each transaction in the dataset.If the total computed from the unit price and quantity does not match the recorded total spent, it indicates a potential issue that could impact financial reporting, inventory tracking, and business analytics along with inconsistency in the dataset.

### **Cell 2: Parameters for checker**

```python

# pre process the specific cols
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Quantity'] = df['Quantity'].astype('Int64')   # quantity can not be a fraction
df['Total Spent']  = pd.to_numeric(df['Total Spent'], errors='coerce')
# df.info()
# menu items prices
expected_menu_prices = {
    "Coffee": 2,
    "Tea": 1.5,
    "Sandwich": 4,
    "Salad": 5,
    "Cake": 3,
    "Cookie": 1,
    "Smoothie": 4,
    "Juice": 3
}
 #Item for which want to check the range
item_to_Check_consistency = 'Tea'
```

### **Cell 3: Checker code**

```python
#func to check consistency
def check_consistency(row, expected_menu, item):
  item_in_row = row['Item']
  # check if the row contains the expected item
  if item_in_row == item:
    expected_item_price = expected_menu.get(item)
    actual_price_per_unit = row['Price Per Unit']
    actual_total_spent = row['Total Spent']

    # check if all data are present and in valid format
    if(pd.notna(row['Quantity']) and pd.notna(row['Price Per Unit'])):
      # check if Quantity and Price Per Unit is positive
      if (row['Quantity'] > 0  and  row['Price Per Unit'] > 0):
        expected_total_spent = row['Quantity'] * row['Price Per Unit']
        # check id actual and exopected total amount matches
        if expected_total_spent != actual_total_spent:
          return f"Expected: {expected_total_spent}, Found: {actual_total_spent}"
      else:
        return f"Quantity or Price Per Unit is either zero or negative"
    else:
      return f"Quantity or Price Per Unit is missing in dataframe"
  else:
    return None

#apply the fuc to dataset for consistency checking
df['Consistency Check'] = df.apply(lambda row: check_consistency(row, expected_menu_prices, item_to_Check_consistency), axis=1)

#filter out the inconsistenyt record
inconsistencies = df[df['Consistency Check'].notna()]

# Display findings
print("Inconsistencies:")
print(inconsistencies[['Transaction ID','Item', 'Quantity', 'Price Per Unit', 'Total Spent', 'Consistency Check']])
if len(inconsistencies) > 0:
  print(f"There are {len(inconsistencies)} inconsistencies for the Item {item_to_Check_consistency}")
else:
  print("No inconsistencies found.")
```

### **Cell 4: Report of findings**
Based on the generated output above, the consistency check has identified 161 transactions for the item "Tea" with discrepancies between the expected and recorded 'Total Spent', primarily due to mismatched calculations where both 'Quantity' and 'Price Per Unit' are present, and due to missing data in either of these fields, which prevents accurate calculation. These inconsistencies, which include several cases where calculated totals differ from recorded totals and instances of missing data (e.g., non-null 'Total Spent' with missing 'Quantity' or 'Price Per Unit'), pose risks to financial reporting and inventory management. In order to conduct an accurate data analysis and gain an accurate insight, a systematic data cleaning process is necessary here.

## Error no:6 (Presence errors)

### **Cell 1: Test Name and description**
  * Error name: **Presence Error**
  * Error description: A "presence error" checker for the "Payment Method" column in the dataset ensures that every entry in this column adheres to the set of predefined valid payment methods. Additionally, it can also be used as pre checker of the fact that all the rows of the column "Payment Method" has been filled with correct values and has not been left empty.

### **Cell 2: Parameters for checker**

```python
# params for presence error
expected_payment_methods = ['Cash', 'Credit Card', 'Debit Card', 'Digital Wallet']
```

### **Cell 3: Checker code**

```python
#fund to check payment method has been selected correctly
def check_payment_method(row, expected_methods):
  payment_method = str(row['Payment Method']).strip()

  if not payment_method:
    return f"Empty payment method: {payment_method}"

  if payment_method not in expected_methods:
    return f"Invalid payment method: {payment_method}"
  return None

#apply the func to the dataset for the payment methid presence check
df['Payment Method Check'] = df.apply(lambda row: check_payment_method(row, expected_payment_methods), axis=1)

#Filter the invalid payment method
invalid_payment_methods = df[df['Payment Method Check'].notna()]

 # Display findings
print("Invalid Payment Methods:")
print(invalid_payment_methods[['Transaction ID', 'Payment Method', 'Payment Method Check']])
if len(invalid_payment_methods) > 0:
  print(f"There are {len(invalid_payment_methods)} invalid payment methods.")
else:
  print("No invalid payment methods found.")

```

### **Cell 1: Test Name and description**

  * Error description: Length check error ensures that an appropriate number of characters are entered into the field. It verifies that neither too long nor too short values is eneterd into the field that can impact the consistencies of the dataset.
  
**Note:** For the purpose of this assignment, we are using Transaction ID to visualize the Length Check validation since no other column in the dataset seemed appropriate for this check. Usually, telephone num, and password fields are used for length check validation; which were not present in the dataset. Therefore, we are using Transaction ID. ### **Cell 2: Parameters for checker**

```python
#param for length check
expected_transaction_id_length = 11
```

### **Cell 3: Checker code**

```python
#func to check transaction ID length
def check_transaction_id_length(row, expected_length):
  transaction_id = str(row['Transaction ID']).strip()
  actual_transaction_id_length = len(transaction_id)
  if actual_transaction_id_length != expected_length:
    return f"Expected length: {expected_length}, Found length: {actual_transaction_id_length}"
  if not transaction_id:
    return f"Empty transaction ID: {transaction_id}"
  return None

#apply the func to each row of the dataset for column Transaction ID
df['Transaction ID Length Check'] = df.apply(lambda row: check_transaction_id_length(row, expected_transaction_id_length), axis=1)

#filter out the invalid length
invalid_transaction_ids = df[df['Transaction ID Length Check'].notna()]

# Display findings
print("Invalid Transaction IDs:")
print(invalid_transaction_ids[['Transaction ID', 'Transaction ID Length Check']])
if len(invalid_transaction_ids) > 0:
  print(f"There are {len(invalid_transaction_ids)} invalid transaction IDs.")
else:
  print("No INVALID transaction IDs found.")
```

According to the generated output of the length validation checker, the transaction IDs in the dataset did not identify any discrepancies. ## Error no: 8 (Look-up errors)

### **Cell 1: Test Name and description**
  * Error name: **Look-up Error**
  * Error description: This error occurs when entries in the 'Location' column do not match the predefined list of accepted location names or are left blank. Invalid or missing location data can lead to inconsistencies in reporting and analysis, affect logistical planning, and potentially disrupt operational efficiencies. Employing a lookup checker as a preliminary check ensures that all entries are either accurate or flagged for correction, thereby preventing any rows from being left empty or incorrectly filled with labels such as 'unknown' or 'error.'r


### **Cell 2: Parameters for checker**

```python
# params for look up error
expected_locations = ['Takeaway','In-store']
```

### **Cell 3: Checker coder**

```python
#func to check look-up errors
def check_location(row, expected_locations):
  location = str(row['Location']).strip()
  if not location:
    return f"Empty location: {location}"
  if location not in expected_locations:
    return f"Invalid location: {location}"
  return None

#apply the func to check the look-up errors
df['Location Look-up Check'] = df.apply(lambda row: check_location(row, expected_locations), axis=1)

#filter out invalid entries from the dataframe
invalid_locations = df[df['Location Look-up Check'].notna()]

# Display findings
print("Invalid Locations:")
print(invalid_locations[['Transaction ID', 'Location', 'Location Look-up Check']])
if len(invalid_locations) > 0:
  print(f"There are {len(invalid_locations)} invalid locations.")
else:
  print("No invalid locations found.")

```

### **Cell 4: report of findings**
From the generated output above we can see that, the dataset's location data reveals a significant number of discrepancies, with 3,961 transactions flagged for having invalid locations. The presence of numerous 'UNKNOWN', 'ERROR', and missing ('NaN') location values necessitates a comprehensive audit and correction initiative.These issues are largely characterized by unrecognized location entries and empty fields. This situation highlights potential flaws in the data entry processes or mismatches in location data synchronization. By using a look-up checker like above, it can be ensured that no invalid data is placed in the dataset.
